```python
#Function for AB test
def ABtest(data, group_col, test_col, alpha=0.05):
    #generate group A 
    A_win = data[data[group_col] == 0][test_col].sum()
    A_loss = len(data[data[group_col] == 0]) - A_win
    A_rate = float(A_win)/(A_win+A_loss)
    #generate group B
    B_win = data[data[group_col] == 1][test_col].sum()
    B_loss = len(data[data[group_col] == 1]) - B_win
    B_rate = float(B_win)/(B_win+B_loss)
    
    #AB chi2 test
    obs = np.array([[A_win,A_loss],[B_win,B_loss]])
    chi2, p, dof, ex = stats.chi2_contingency(obs)    

    #Print results if test result is statistically significant
    if p < alpha:
        print ''
        print group_col, 'Group A', test_col, A_win, 'non',test_col, A_loss, test_col,'rate', A_rate*100, '%'
        print group_col, 'Group B', test_col, B_win, 'non',test_col, B_loss, test_col,'rate', B_rate*100, '%'
        print ''
        print 'Accept population difference by {}% confidence level.' .format((1-p)*100)
        if (B_rate-A_rate) > 0:
            return '{} Group B has a higher {} rate of {}% than Group A.' .format(group_col, test_col, round((B_rate-A_rate)*100,2))
        elif (B_rate-A_rate) < 0:
            return '{} Group B has a lower {} rate of {}% than Group A.' .format(group_col, test_col, round((B_rate-A_rate)*100,2))
```

```python
#Group high customer service calls (>1) as high service demand customers
df['Service_demand'] = (df['Customer service calls'] > 1)*1
ABtest(df, 'Service_demand', 'Churn')
```

```python
#Group high customer service calls (>1) as high service demand customers
df['High_day_charge'] = (df['Total day charge'] > df['Total day charge'].mean())*1
ABtest(df, 'High_day_charge', 'Churn')
```

```python
def bootstrap(sample, stat_func, iters=1000):
    boots = []
    for i in range(iters):
        random_sample = np.random.choice(sample, replace=True, size=len(sample))
        stat = stat_func(random_sample)
        boots.append(stat)
    return boots
```

```python
servicechurn_boot = bootstrap(Service_dd_churn, np.mean)

servicechurn_lower = stats.scoreatpercentile(servicechurn_boot, 2.5)
servicechurn_upper = stats.scoreatpercentile(servicechurn_boot, 97.5)

print servicechurn_lower, np.mean(Service_dd_churn), servicechurn_upper
```

```python
daychar_boot = bootstrap(daychar_dd_churn, np.mean)

daychar_lower = stats.scoreatpercentile(daychar_boot, 2.5)
daychar_upper = stats.scoreatpercentile(daychar_boot, 97.5)

print daychar_lower, np.mean(daychar_boot), daychar_upper
```

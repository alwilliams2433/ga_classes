```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
movies = pd.read_csv("./datasets/movie_weekend.csv")
```

```python
movies.info()
#27 null values
```

```python
movies[movies['MOVIE'].isnull()].head(3)
#null values are empty rows, can be deleted
```

```python
movies = movies.dropna()
movies.info()
```

```python
#spider-man 4,5,6,7, belongs to spider-man3, to be renamed
movies[movies['MOVIE'].isin(['Spider-Man 3','Spider-Man 4','Spider-Man 5','Spider-Man 6','Spider-Man 7'])]
```

```python
movies.loc[movies['MOVIE'].isin(['Spider-Man 4','Spider-Man 5','Spider-Man 6','Spider-Man 7']), ['MOVIE']] = 'Spider-Man 3'
```

```python
#Which movies remained in the longest?
longest = movies.groupby('MOVIE').max().sort_values('WEEK_NUM', ascending=False)['WEEK_NUM'].head(10)
longest.sort_values(ascending=True).plot(kind='barh', figsize=(15,8))
plt.savefig('longest.png')
```

```python
#Which movies were good investments?
df = movies.groupby('MOVIE').sum().sort_values('WEEKEND_PER_THEATER', ascending=False)['WEEKEND_PER_THEATER'].head(10)
movies.groupby('MOVIE').sum().sort_values('WEEKEND_PER_THEATER', ascending=False)['WEEKEND_PER_THEATER'].head(10)
```

```python
receipts = movies.groupby('MOVIE').sum()['WEEKEND_PER_THEATER']
weeks = movies.groupby('MOVIE').max()['WEEK_NUM']
receiptperweek = receipts/weeks
```

```python
testdf = pd.concat([receiptperweek,weeks],axis=1,join="inner")
testdf.columns=['Receipt_per_week','Running_weeks']
testdf.plot.scatter('Receipt_per_week','Running_weeks', figsize=(15,8))
```

```python
avgreceipt = receiptperweek.sort_values(ascending=False).head(10)
avgreceipt.sort_values(ascending=True).plot(kind='barh', figsize=(15,8))
plt.savefig('average.png')
```

```python
movies.groupby(['WEEK_NUM','MOVIE']).sum()['WEEKEND_PER_THEATER'].unstack().plot(figsize=(15,8))
plt.savefig('receiptsoverweeks.png')
```

```python
movies['WEEKEND_DATE'] = pd.to_datetime(movies['WEEKEND_DATE'])
movies['Month'] = movies['WEEKEND_DATE'].dt.month
```

```python
#overall trend of weekend receipts per movie over the years
movies.groupby('WEEKEND_DATE').mean().sort_index()['WEEKEND_PER_THEATER'].plot(figsize=(15,8))
```

```python
#monthly trend of receipt per movie
receipts_month = movies.groupby('Month').sum()['WEEKEND_PER_THEATER'].sort_index()
movies_month = movies.groupby('Month')['MOVIE'].unique().map(len)

(receipts_month/movies_month).plot(figsize=(15,8)).set_ylabel('Earnings per Movie')
plt.savefig('seasonal.png')
```

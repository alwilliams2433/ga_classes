#### 4. What is the observation count per shape?

```python
df.groupby('Shape Reported').size().sort_values(ascending=False)
```

#### 5. Create a subset of the data that contains the top five cities and the top five shapes.

```python
topshape = df.groupby('Shape Reported').size().sort_values(ascending=False).head(5)
topcity =  df.groupby('City').size().sort_values(ascending=False).head(5)
```

```python
print topshape.index.values.tolist()
print topcity.index.values.tolist()
cities = topcity.index.values.tolist()
shapes = topshape.index.values.tolist()
```
